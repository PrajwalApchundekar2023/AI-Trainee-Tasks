##**Random Forest Algorithm**
This Technique uses Both classification and regresion
1. Bootstrap / Bootstraping Sampling.
2. Voting and avaraging
3. it use desion tree construction




```python
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
dir(digits)
```

```python
print(digits.target_names)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(digits.data)
```

```python
df.head()
```

```python
df['target'] = digits.target
```

```python
print(digits.target)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop(['target'],axis='columns'),digits.target,test_size=0.9)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
model = RandomForestClassifier(n_estimators=40)
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
y_predicted = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predicted)
cm
```

```python
import matplotlib.pyplot as plt
import seaborn as sn
sn.heatmap(cm, annot=True)
plt.xlabel('predicted')
plt.ylabel('True')
```
